```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X=dataset.iloc[:,[3,4]].values
X
```

```python
from sklearn.cluster import Birch  

Birch = Birch(threshold=0.5, branching_factor=50, n_clusters=5, compute_labels=True, copy=True)
y_bcl=Birch.fit_predict(X)
```

```python
y_bcl
```

```python
supervised=dataset
supervised
```

```python
supervised['Cluster_group']=y_bcl
```

```python
supervised
```

```python
supervised.to_csv("BirchClustering.csv",index=False)
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```

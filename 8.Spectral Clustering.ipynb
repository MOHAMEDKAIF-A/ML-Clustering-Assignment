```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X=dataset.iloc[:,[3,4]].values
X
```

```python
from sklearn.cluster import SpectralClustering  


spectralcl = SpectralClustering(n_clusters=5, eigen_solver=None, random_state=None, n_init=30, gamma=5.0)
y_spectralcl=spectralcl.fit_predict(X)
```

```python
y_spectralcl
```

```python
supervised=dataset
supervised
```

```python
supervised['Cluster_group']=y_spectralcl
```

```python
supervised
```

```python
supervised.to_csv("SpectralClustering.csv",index=False)
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X=dataset.iloc[:,[3,4]].values
X
```

```python
from sklearn.cluster import AffinityPropagation  
from sklearn.datasets import make_blobs

apcl = AffinityPropagation(damping=0.6, max_iter=200, convergence_iter=15, copy=True, preference=None, affinity='euclidean', verbose=False,)
y_apcl = apcl.fit_predict(X)  
```

```python
y_apcl
```

```python
supervised=dataset
supervised
```

```python
supervised['Cluster_group']=y_apcl
```

```python
supervised
```

```python
supervised.to_csv("affinitypropogation.csv",index=False)
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```
